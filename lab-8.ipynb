# 8. Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates. 

```python
import pandas as pd
import matplotlib.pyplot as plt

#Read Dataset
data = pd.read_csv('/kaggle/input/stock-market-data-nifty-50-2000-2022/Nifty_50_with_indicators_.csv');

# Column of Date
data['Date'] = pd.to_datetime(data['Date'])

#Start date and End date
start_date = '2000-01-03'
end_date = '2022-06-09'

date = data[(data['Date'] >= start_date) & (data['Date'] <= end_date)]

#Create a Plot
plt.figure(figsize=(12,6))

plt.plot(data['Date'], data['open'], label='Open Price', marker='o', linestyle='-')
plt.plot(data['Date'], data['close'], label='Close Price', marker='o', linestyle='-')

plt.title('Stock Prices Between {} and {}'.format(start_date, end_date))
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
