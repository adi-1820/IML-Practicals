# 9. Write a Pandas program to create a plot of Open, High, Low, Close, Adjusted Closing prices and Volume of given company between two specific dates.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Dataset
data = pd.read_csv('/kaggle/input/stock-market-analysis-data/stocks.csv')

# Coulmn of ate
data['Date'] = pd.to_datetime(data['Date'])

#Start Date and End Date
start_date = '2023-02-07'
end_date = '2023-05-05'

date = data[(data['Date'] >= start_date) & (data['Date'] <= end_date)]

# Create a Plot
plt.figure(figsize=(12,6))

plt.plot(date['Date'], date['Open'], label='Open', linewidth=2)
plt.plot(date['Date'], date['High'], label='High', linewidth=2)
plt.plot(date['Date'], date['Low'], label='Low', linewidth=2)
plt.plot(date['Date'], date['Close'], label='Close', linewidth=2)
plt.plot(date['Date'], date['Adj Close'], label='Adj Close', linewidth=2)

plt.bar(date['Date'], date['Volume'], label='Volume', alpha=0.5)

# Customize the plot
plt.title('Stock Price and Volume Analysis')
plt.xlabel('Date')
plt.ylabel('Price / Volume')
plt.legend()
plt.grid(True)
plt.show()
```
