# 9. Write a Pandas program to create a plot of Open, High, Low, Close, Adjusted Closing prices and Volume of given company between two specific dates.

```python
# 2162070307032 - Aditya Vara
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go

# Read Dataset
data = pd.read_csv('/kaggle/input/stock-market-analysis-data/stocks.csv')

# Coulmn of ate
data['Date'] = pd.to_datetime(data['Date'])

#Start Date and End Date
start_date = '2023-02-07'
end_date = '2023-05-05'

filtered_data = data[(data['Date'] >= start_date) & (data['Date'] <= end_date)]

# Create a Figure
fig = go.Figure(data=[go.Candlestick(x=data['Date'],
                                     
                                     open=data['Open'],
                                     high=data['High'],
                                     low=data['Low'],
                                     close=data['Close']
                                    )])

# Show the Figure
fig.show()
```
